**Converting EndTime to DateTime Format & Extracting Specific Components**

```python
stream_pd['endTime'] = pd.to_datetime(stream_pd['endTime'])
```

```python
stream_pd['endTime']
```

```python
stream_pd.dtypes
```

```python
stream_pd['date']  = stream_pd['endTime'].dt.date
stream_pd['hour']  = stream_pd['endTime'].dt.hour
stream_pd['weekday'] = stream_pd['endTime'].dt.day_name()
stream_pd['month']   = stream_pd['endTime'].dt.month_name()


print(stream_pd[['endTime', 'date', 'hour', 'weekday', 'month']].head())
```

**Identifying Skipped Songs**

```python
stream_pd['minutes_played'] = (stream_pd['msPlayed']/60000).round(2)
```

```python
stream_pd['minutes_played'].head()
```

```python
stream_pd ['skipped_songs'] = stream_pd['msPlayed'] < 30000
```

```python
stream_pd ['skipped_songs'].head(100)
```
